```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
battles = pd.read_csv('dataset/battles_cleaned.csv')
```

```python
pd.set_option('display.max_columns', None)
battles
```

### List Pertanyaan
- Attacker King yang memiliki pasukan terbanyak
- Defender King yang memiliki pasukan terbanyak
- King yang paling sering menyerang
- King yang paling sering diserang
- King yang paling banyak memenangi peperangan
- Jenis Perang yang paling banyak dimenangi oleh setiap King
- King yang memiliki Commander terbanyak (setelah jumlah commander dihitung)
- Hubungan antara jumlah Commander dengan kemenangan (setelah jumlah commander dihitung)
- Hubungan antara kemenangan perang seorang Raja dengan lokasi perang
- Hubungan antara kemenangan perang seorang Raja dengan Jenis Perang
- Hubungan antara kemenangan perang seorang Raja dengan jumlah pasukan
- Hubungan antara kemenangan perang House of Stark dengan musim panas

```python
df = battles.copy()
```

```python
df['attacker_size'].groupby(df['attacker_king']).sum().plot(kind='barh')
plt.title("Attacker Army Size") 
plt.grid()
plt.show()
```

```python
df['attacker_king'].value_counts().sort_values(ascending=True).plot(kind='barh')
plt.title("Most Aggressive King") 
plt.grid()
plt.show()
```

```python
df['defender_king'].value_counts().sort_values(ascending=True).plot(kind='barh')
plt.title("Most Defensive King") 
plt.grid()
plt.show()
```

```python
outcome = battles['attacker_outcome'].groupby(battles['attacker_king']).value_counts()
outcome.unstack(level='attacker_outcome').plot(kind='barh')
plt.title("Battle Outcome") 
plt.grid()
plt.show()
```

```python
win = df[df['attacker_outcome']=='win']
outcome = win['battle_type'].groupby(win['attacker_king']).value_counts()
outcome.unstack(level='battle_type').plot(kind='barh',figsize=(10,6))
plt.title("The type of war that each King won the most") 
plt.show()
```
